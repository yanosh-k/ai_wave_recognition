```python
from fastai.vision.all import *
import os
```

```python
def is_waving(fileName):
    return not fileName[0].isupper()
```

```python
learn = load_learner(Path(os.getcwd()) / 'model_v3.pkl')
```

```python
import cv2
from IPython.display import clear_output, display
from termcolor import colored
import time
```

```python
cap = cv2.VideoCapture(2)
cap.set(3,640) # adjust width
cap.set(4,480) # adjust height

imgCounter = 0

while True:
    # Read image from camera
    success, img = cap.read()
    flippedImg = cv2.flip(img, 1)

    # Convert the cv2 image to PILImage
    img = cv2.cvtColor(flippedImg, cv2.COLOR_BGR2RGB)
    imgPil = PILImage.create(img)

    # Infer the waving
    is_waving, _, probs = learn.predict(imgPil)

    imgText = 'WAVING' if is_waving == 'True' else 'NOT WAVING'
    imgTextColor = (0, 255, 0) if is_waving == 'True' else (0, 0, 255) # green if waving, red otherwise (in BGR)

    
    cv2.putText(flippedImg, imgText + f': {probs[1].item():.6f}', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, imgTextColor, 2, cv2.LINE_AA)
    
    # Show the wavign capture window
    cv2.imshow("Webcam", flippedImg) # This will open an independent window

    # Close window if the 'q' key is pressed
    key = cv2.waitKey(1)
    if key & 0xFF == ord('q'): # quit when 'q' is pressed
        cap.release()
        break 

cv2.destroyAllWindows() 
```
